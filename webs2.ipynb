```python
import requests
from bs4 import BeautifulSoup

# URL of the Goodreads list
url = 'https://www.goodreads.com/list/show/1.Best_Books_Ever'

# Headers to mimic a browser request
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
}

# Send a GET request to the URL with headers
response = requests.get(url, headers=headers)

# Check if the request was successful
if response.status_code == 200:
    # Parse the HTML content of the page
    soup = BeautifulSoup(response.content, 'html.parser')
    
    # Find all the book title elements
    book_titles = soup.find_all('a', class_='bookTitle')
    book_authors = soup.find_all('a', class_ = 'authorName')
    book_ratings = soup.find( 'a', class_ = 'minirating')

    
    # Extract and print the titles of the books
    t_count = 1
    for book_t in book_titles:
       
        title = book_t.find('span').text
        print(title, t_count)
        t_count = t_count+1

     # Extract and print the authors of the books
    a_count = 1
    for book_a in book_authors:
         
         authors= book_a.find('span').text
         print(authors, a_count)
         a_count = a_count +1

    # Extract and print the ratings of the books
    r_count = 1
    for book_r in book_ratings:
         
         ratings= book_r.find('span').text
         print(ratings, r_count)
         r_count = r_count +1
    
else:
    print(f'Failed to retrieve the webpage. Status code: {response.status_code}')
```
